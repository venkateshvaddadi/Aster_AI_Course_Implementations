**Read the video and converting into the images**

[link text](https://)

1.   We need to read the video
2.   Make the folder for the each specific video
3.   Make the video into frames
4.   save the each frame of the video as image in their respective folder.
5.   make a enrty for of the generated image path and respective label.
6.   repeat this for the each video.



```python
videos_path=data_path+'/pocus_videos/'
print(os.listdir(videos_path))
import cv2

for i in os.listdir(videos_path):
    for label in os.listdir(videos_path+'/'+i):
        # print(label)
        for video_name in os.listdir(videos_path+'/'+i+'/'+label):
            video_path=videos_path+'/'+i+'/'+label+'/'+video_name
            video_label=label
            #print(video_path,video_label)
            # need to write a code here to convert the video into images
            # video can be .avi, .mpeg ,mp4 , .gif,.mov
            if(video_name.lower().endswith('.mov') ):
                print(video_path)
                video_to_folder_name = os.path.splitext(os.path.basename(video_name))[0]
                print(video_name)
                try:
                    video_to_folder_full_path=data_path+'pocus_videos_to_images_updated/'+i+'/'+label+'/'+video_to_folder_name
                    os.makedirs(video_to_folder_full_path)
                except:
                    print('folder already existed')
                    # code for converting video into images


                video = cv2.VideoCapture(video_path)
                frame_count = 0
                # Read frames from the video and save them as images
                while True:
                    ret, frame = video.read()
                    if not ret:
                        break
                    print(frame.shape)

                    # Save the frame as a PNG image
                    generated_file_name=video_to_folder_full_path+"/frame_"+str(frame_count)+".png"

                    # write the image on target folder
                    cv2.imwrite(generated_file_name, frame)
                    frame_count += 1

                    # saving the generated image paths and their labels.
                    image_paths.append(generated_file_name)
                    image_labels.append(label)

                # Release the video capture object
                video.release()

```
